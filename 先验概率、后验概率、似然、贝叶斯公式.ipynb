## 1.“概率”与“似然”  
“概率论”和“数理统计”其实是一对相反的过程。
* “概率论”是我们已经有个概率模型，模型的参数都有了，我们来研究它产生的数据的性质(例如均值等)。即：有模型，有参数->求数据  
* “数理统计”是我们有概率模型，有数据，但是没有模型参数，我们通过数据来算参数。即：有模型，有数据->求参数  

对于式子$P(x|θ)$:
* 在概率论中，这个式子叫做“概率”。参数θ是已知的，数据x是未知的  
* 在数理统计中，这个式子叫“似然”。参数θ是未知的，数据x是已知的  

例：抛硬币，P(x|θ)=P(x=正面)=θ  
* 概率论的角度：我们【假设】硬币是均匀的，于是我们有θ=1/2，于是我们可以计算抛十次硬币得到某个正反面结果序列(例：+++-++-+--)  
* 统计的角度：我们不知道这个硬币是否均匀，于是我们做了个抛硬币实验，得到一个结果序列(例：+++-++-+--)。有这个序列我们估计(例如用极大似然估计法)求出θ=4/10。之后再预测硬币序列时，我们带入估计得出的参数θ来计算  

抛十次硬币全是正面，下一次猜硬币猜是正面还是反面：
* 概率的角度：如果我们假设硬币是均匀的，每次正反都是1/2，则每次的结果都是独立的，和之前观测的数据无关，下一次无论猜正反，猜对的比例都是1/2，和前面的无关  
* 统计的角度：我们认为硬币有可能是不均匀的。前十次我们估计出硬币是不均匀的，下一次猜正面猜对的比例更大  

## 2.先验，后验，似然，贝叶斯公式
贝叶斯公式：$P(A_i|B)=\frac{P(B|A)P(A_i)}{\sum_iP(B|A_i)P(A_i)}$=$\frac{P(B|A)}{\sum_iP(B|A_i)P(A_i)}P(A_i)$  
* $P(A_i)$: $A_i$的先验概率  
* $P(A_i|B)$: 在观察到$B$时$A_i$的后验概率  
* $\frac{P(B|A)}{\sum_iP(B|A_i)P(A_i)}$:证据$B$的似然  

**例：垃圾邮件**  
事件$A_1$:邮件是垃圾邮件,事件$A_2$:邮件是正常邮件。事件$B$:邮件提到了“海景房”  
我们有一堆邮件。从中可以直接统计出垃圾邮件的比例，这个比例就是先验概率$P(A_1)$：不依赖于对邮件内容等观察结果，直接盲猜，猜中一个邮件是垃圾邮件的概率。  
我们再看看邮件里的内容。统计出垃圾邮件中提到“海景房”的概率$P(B|A_1)$ ，正常邮件提到海景房的概率$P(B|A_2)$。由贝叶斯公式可以算出，一封提到“海景房”的邮件，它是垃圾邮件的概率$P(A_1|B)$。这就是后验概率$P(A_1|B)$：在观察到了证据$B$：邮件提到了“海景房”的情况下，猜邮件是垃圾邮件猜对的概率  
B的似然：有了证据B对我们的猜想A带来的可靠性的提升：

先验概率:没有任何证据时，盲猜能猜对的概率（先验概率可以是我们统计来的，也可以是我们凭感觉、凭经验给一个值）  
证据的似然:对证据B进行统计，从而得出证据B对猜想A的支持力度（是统计得到的）  
后验概率:在拿到证据B时，能猜对的概率（是通过贝叶斯公式计算得到的）  
贝叶斯公式的意义：我们有了先验概率，也有了关于证据B的一堆统计结果，就可以算出证据B的似然，用其修正A的先验概率，得出在有证据B佐证的条件下A的后验概率  

**另一个例子：疾病假阳性**
某疾病发病率0.1%,检查正确率99%(即：正常人99%概率检测出阴性，病人99%检测出阳性)  
P(病|阳性) = $\frac{P(阳|病)}{P(阳|病)P(病)+P(阳|没病)P(没病)}P(病)=\frac{0.99}{0.99*0.001+0.01*0.999}*0.001=90.16 * 0.001=0.09$。  
似然=90.16>1，即阳性结果增大了我们猜一个人是病人的猜对概率。后验概率0.09大于先验概率0.001.但是这个后验概率依然不太大，还不足以让我们相信一个人是病人  
P(没病|阳性) = $\frac{P(阳|没病)}{P(阳|病)P(病)+P(阳|没病)P(没病)}P(没病)=\frac{0.01}{0.99*0.001+0.01*0.999}*0.001=0.9 * 0.999=0.91$  
似然0.9<1，级阳性结果减小了我们猜一个人是健康人的概率。但是后验概率依然很大，阳性不足以让我们否认一个人是健康人  
**注：先验概率和为1：P(病)+P(没病)=1.后验概率和也为1，P(病|阳性)+P(没病|阳性)=1. 但是似然的和不为1。每一个似然都允许大于1、小于1、等于1.似然大于1：证据增强了我们先验概率的猜想。小于1：证据减弱了先验概率的猜想**  
**注2：**这里的似然，与前面提到的“1.概率与似然”的“似然”说的不是一个事。“1.概率与似然”说的是似然函数$P(x|θ)$。这里的似然指的是一个证据对先验概率的支持力度。但相同点在于，两个“似然”都是通过统计得出来的。另外，在下面的“极大似然估计”、“极大后验估计”中，我们会看到，两个“似然”其实确实就是一回事
